# Displaying bar charts of discrete distributions
    JMA 29 Jun 2026

```python
import matplotlib.pyplot as plt

distribution1 = [0.4, 0.6]

distribution2 = [0.15, 0.5, 0.35]
```

```python
# Create side-by-side subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# ---- Left: Vertical Bar Chart ----
x1 = range(len(distribution1))
x2 = range(len(distribution2))
width = 0.35

ax1.bar(x1, distribution1, width, color='steelblue', alpha=0.8, label='Event 1')
# ax1.bar([i + width for i in x2], distribution2, width, color='coral', alpha=0.8, label='Event 2')
ax1.set_xlabel('Index')
ax1.set_ylabel('Probability')
ax1.set_title('Vertical Bar Chart')
ax1.set_xticks([i + width/2 for i in x1])
ax1.set_xticklabels(range(len(distribution1)))
ax1.legend()
ax1.set_ylim(0, 0.7)

# ---- Right: Horizontal Bar Chart ----
y1 = range(len(distribution1))
y2 = range(len(distribution2))
height = 0.35

# ax2.barh(y1, distribution1, height, color='steelblue', alpha=0.8, label='Event 1')
ax2.barh([i + height for i in y2], distribution2, height, color='coral', alpha=0.8, label='Event 2')
ax2.set_ylabel('Index')
ax2.set_xlabel('Probability')
ax2.set_title('Horizontal Bar Chart')
ax2.set_yticks([i + height/2 for i in y2])
ax2.set_yticklabels(range(len(distribution2)))
ax2.legend()
ax2.set_xlim(0, 0.7)

plt.tight_layout()
plt.show()
```
