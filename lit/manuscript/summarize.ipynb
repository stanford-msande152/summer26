RUn regular expressions over lines in markdown to summarize the document outline.

To extract visual content from a PDF tell a model to use pymuPDF (import fitz) 

```python
import re, os, sys
CNT_LIMIT = 1000000
```

```python
# re - section lines

re_section = r'^\#+'
re_figure = r'^\*Figure.+\*$'
```

```python
with open('MS&E152_coursenotes.md', 'r') as cn, open('MS&E_table_of_contents.md', 'wt') as sm:
    for k, ln in enumerate(cn):
        ln = ln.strip()
        figure_re = re.search(re_figure, ln)
        header_re = re.search(re_section, ln)
        if (header_re is not None):
            ln_hdr = len(ln) 
            tbs = (119 - ln_hdr)//4
            sm.write(ln + tbs*'\t' + f':{k+1}\n')
        if (figure_re is not None):
            sm.write(f'\t{ln}\n')
        if k > CNT_LIMIT:
            break
```
