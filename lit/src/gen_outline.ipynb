Outline the document by running regular expressions over lines in markdown.


```python
import re, os, sys
CNT_LIMIT = 1000000
```

```python
# re - section lines

re_section = r'^\#+'
re_figure = r'^\*Figure.+\*$'

src_file = 'Howard_manuscript.md'
```

```python
# if len(sys.argv) > 1:
#     src_file = sys.argv[1].strip()
if not re.search(r'\.md$', src_file):
    print(src_file, ' is not a markdown file')
    assert False        # Stop execution of subsequent cells
dst_file = src_file.replace(r'.md', r'_toc.md')  # Note replace doesn't take regular expressions. 
dst_file
```

```python
with open(src_file, 'r') as cn, open(dst_file, 'wt') as sm:
    for k, ln in enumerate(cn):
        ln = ln.strip()
        figure_re = re.search(re_figure, ln)
        header_re = re.search(re_section, ln)
        if (header_re is not None):
            ln_hdr = len(ln) 
            tbs = (119 - ln_hdr)//4
            sm.write(ln + tbs*'\t' + f':{k+1}\n')
        if (figure_re is not None):
            sm.write(f'\t{ln}\n')
        if k > CNT_LIMIT:
            break
```
